```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

df = pd.read_csv("aep_data.csv")
print(df.head())


```

```python
df.dropna(subset=["PNT_ATRISKNOTES_TX"], inplace = True)
df.drop_duplicates(inplace = True)
print(df.head())
```

```python
df['datetime_parsed'] = pd.to_datetime(df['DATETIME_DTM'], errors='coerce')
df['date'] = df['datetime_parsed'].dt.date
df['time'] = df['datetime_parsed'].dt.time

print(df.head())
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    # convert lowercase
    text = text.lower()
    # remove special characters
    text = re.sub(r'\W+', ' ', text)
    # tokenize
    words = word_tokenize(text)
    # remove stopwords
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
# apply to 'PNT_ATRISKNOTES_TX' and 'QUALIFIER_TXT'
df['cleaned_atrisk_notes'] = df['PNT_ATRISKNOTES_TX'].apply(preprocess_text)
df['cleaned_qualifier'] = df['QUALIFIER_TXT'].apply(preprocess_text)

print(df.head())
```

```python
tfidf = TfidfVectorizer(max_features=1000)
new_tfidf = tfidf.fit_transform(df['cleaned_atrisk_notes'])
print(df.head())
print(new_tfidf)
```

```python
df.to_csv('cleaned_safety_observations.csv', index=False)

cdf = pd.read_csv('cleaned_safety_observations.csv')
cdf.drop(columns = ['QUALIFIER_TXT', 'DATETIME_DTM', 'PNT_ATRISKNOTES_TX'], inplace = True)
cdf.to_csv('cleaned_data.csv')
print(cdf.head())
```
